```python
import pandas as pd
```

```python
df = pd.read_csv ('inputs/09.csv',sep=' ',header=None)
df = df.astype("str")
df=df[0].str.split("",expand=True).copy()
df=df.drop(columns=[0,len(df.columns)-1])
df.head(5)
```

## 09/01

```python
def is_smaller(value,x,y,df):
    if x > 0 and x < len(df.columns)+1 and y >= 0 and y < len(df.index):
        return value < df.loc[y][x]
    return True

df01 = df.copy()

low_points = {}

for r in range(len(df01.index)):
    for c in range(1,len(df01.columns)+1):
        if is_smaller(df.loc[r][c],c,r-1,df01) and is_smaller(df.loc[r][c],c-1,r,df01) and is_smaller(df.loc[r][c],c,r+1,df01) and is_smaller(df.loc[r][c],c+1,r,df01):
            low_points[(c,r)] = df.loc[r][c]

low_points
```

```python
pd.DataFrame(low_points,index=[0]).astype("int32").apply(lambda x: x+1).iloc[0].sum()
```

## 09/02

```python
def get_basin(basin,r,c,df):
    #print(basin)
    #print(r,c)

    if df.iloc[r][c] != '9' and (c,r) not in basin:
        basin[(c,r)] = df.iloc[r][c] 

        if c-1 > 0:
            basin = get_basin(basin,r,c-1,df)
        if c+1 < len(df.columns)+1:
            basin = get_basin(basin,r,c+1,df)
        if r-1 >= 0:
            basin = get_basin(basin,r-1,c,df)
        if r+1 < len(df.index):
            basin = get_basin(basin,r+1,c,df)
    
    return basin

df01 = df.copy()

basins = []

for r in range(len(df01.index)):
   for c in range(1,len(df01.columns)+1):
       if df01.iloc[r][c] < '9':
        basin = {}
        basins.append(get_basin(basin,r,c,df01))
        for x in basin:
            df01.iloc[x[1]][x[0]] = '9'
basins
```

```python
basin_sizes = []
for x in basins:
    basin_sizes.append(len(x))
basin_sizes.sort(reverse=True)
basin_sizes
```

```python
basin_sizes[0]*basin_sizes[1]*basin_sizes[2]
```
