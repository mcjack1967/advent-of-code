```python
import pandas as pd
```

```python
df = pd.read_csv ('inputs/08.csv',sep=' ',header=None)
df.head(10)
```

```python
# define digits
digits = { 0: 'abcefg', 1: 'cf', 2:'acdeg', 3:'acdfg', 4:'bcdf'
            , 5:'abdfg', 6:'abdefg', 7:'acf', 8:'abcdefg', 9:'abcdfg'
         }

dfdig = pd.DataFrame.from_dict(digits, orient='index', columns=["pattern"])
dfdig["length"] = dfdig.apply(lambda row : len(row["pattern"]), axis=1)
dfdig
```

```python
dftemp = dfdig.groupby("length").count().reset_index()
dftemp[dftemp["pattern"]==1]["length"].values
dftemp = dfdig[dfdig["length"].isin(dftemp[dftemp["pattern"]==1]["length"].values)].copy()
dftemp
```

```python
dfout = df[list(range(11,df.columns.size))].copy()
dfout
```

```python
def get_lenght(list):
    #print(list)
    out = []
    for x in list:
        out.append(len(x))
    return pd.Series(out)

def get_number_with_lenth(list,lengths):
    out = []
    for x in list:
        out.append(1 if len(x) in lengths.values else 0)
    return pd.Series(out)

"Number of unique output numbers are {}".format(dfout.apply(get_number_with_lenth,args=[dftemp["length"]]).sum().sum())
```

```python
letters = ["a","b","c","d","e","f","g"]

for l in letters:
    dfdig[l] = dfdig.apply(lambda row : 1 if l in row["pattern"] else 0, axis=1)
dfdig["number"]=dfdig.index
dfdig#[letters]
```

```python
def get_numbers(list,dflen):
    out = []
    pattern = []
    lenght = []
    for x in list:
        pattern.append("".join(sorted(x)))
        out.append(dflen[dflen["length"]==len(x)].index.tolist()[0] if len(x) in dflen["length"].values else -1)
        lenght.append(len(x))
    df = pd.concat([pd.Series(pattern),pd.Series(out),pd.Series(lenght)],ignore_index=True)

    return df

def create_df(df,i):
    df = df.iloc[i:i+1].copy()
    df=df[list(range(10))].apply(get_numbers,args=[dftemp]).T
    df = df.rename({0:"pattern",1:"number",2:"length"},axis=1)

    for l in letters:
        df[l] = df.apply(lambda row : 1 if l in row["pattern"] else 0, axis=1)
    return df
```

```python
def solve(df,mapping,a,n):
    for x in df[df["number"]==n]["pattern"].values[0]:
        #print("x:",x)
        list = mapping[a]
        if x in list:
            mapping[a] = x

    for i in mapping:
        #print(type(mapping[i]))
        if type(mapping[i]) is not str:
            if mapping[a] in mapping[i]:
                mapping[i].remove(mapping[a])
                mapping[i] = "".join(mapping[i])
    return mapping

def mapping(dfdig,df):
    mapping = {}

    for i in dfdig[letters].sum().index:
        for j in df[letters].sum().index:
            if dfdig[letters].sum()[i] == df[letters].sum()[j]:
                if i in mapping:
                    mapping[i] = [mapping[i],j]
                else:
                    mapping[i] = j

    mapping = solve(df,mapping,"c",1)
    mapping = solve(df,mapping,"d",4)

    return mapping
```

```python
total = 0

for i in range(0,len(df.index)):
    df01 = create_df(df,i).copy()

    df01 = pd.merge(df01,dfdig.rename(columns=mapping(dfdig,df01)),how="left",on=letters)[["pattern_x","number_y"]]

    df02=df.iloc[i:i+1][list(range(11,15))]
    df02=df02.apply(get_numbers,args=[dftemp]).T

    str_list = []
    for n in pd.merge(df02,df01,how="left",left_on=0,right_on="pattern_x")["number_y"].values:
        str_list.append(str(n))
    total = total + int("".join(str_list))
```
